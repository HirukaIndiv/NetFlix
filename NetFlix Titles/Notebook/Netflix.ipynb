### 2.2 Growth of content from India, Korea, Japan, Europe, etc.

```python
content_change = netflix.groupby('country')['release_year'].value_counts().groupby(level=0).pct_change().fillna(0)
content_change = content_change.reset_index(name='content_change')
content_change
```

```python
content_df= content_change.assign(country=content_change['country'].str.split(', ')).explode('country')
content_df
```

```python
content_df['country'] = content_df['country'].str.strip()
content_df
```

```python
netflix = netflix.reset_index(drop=True)
content_df = content_df.reset_index(drop=True)

netflix['country'] = content_df['country']
netflix['country'] = content_df['country'].reset_index(drop=True)
```

```python
netflix.to_csv('netflix.csv', index=False)
```

```python
content_change=(content_df.groupby('country')['release_year'].value_counts().groupby(level=0).pct_change().fillna(0))
content_change
```

```python
content_growth = content_change.reset_index(name='change').sort_values(by='change',ascending=True)
content_growth
```

```python
content_growth = (netflix.groupby(['release_year', 'country']).size().pct_change())
```

```python
content_growth_df = content_growth.reset_index(name='growth').sort_values(['country', 'growth'])
content_growth_df
```

# 3. TALENT STRATEGY (DIRECTORS AND CAST)

### 3.1 Most frequent directors

```python
freq_director=netflix['director'].value_counts().drop('Not Known')
freq_director
```

```python
most_freq_director_df = freq_director.to_frame().reset_index()
```

```python
most_freq_director_df.columns = ['Director', 'count']
most_freq_director_df
```

### 3.2 Most frequent actor

```python
actors = (netflix.assign(actor=netflix['cast'].str.split(', ')).explode('actor'))
actors
```

```python
actors_df = actors[(actors['actor'].str.strip() != '') & (actors['actor'].str.strip() != 'No Idea')]
most_freq_actor = actors_df['actor'].value_counts().head(10)
most_freq_actor
```

```python
most_freq_actor_df = most_freq_actor.to_frame().reset_index()
```

```python
most_freq_actor_df.columns = ['actor', 'counts']
most_freq_actor_df
```

```python
combo = pd.merge(most_freq_actor_df,most_freq_director_df, how='cross')
combo
```

```python
combination = (netflix.groupby(['cast', 'director']).size().sort_values(ascending=False)).to_frame()
combination
```

```python
import json

# Load the JSON data
input_file = 'world.json'

with open(input_file, 'r') as f:
    data = json.load(f)

# --- Step 1: Target the 'land' geometry list ---
land_geometries = data['objects']['land']['geometries']
initial_land_count = len(land_geometries)

# --- Step 2: Remove the last large feature (Antarctica) ---
# Based on the file structure (where Antarctica is drawn as the last large polygon),
# removing the last element in the list is often the simplest fix.
if land_geometries:
    # Delete the last item in the list of land geometries
    del land_geometries[-1] 

# --- Step 3: Save the modified data to a new file ---
output_file = 'worlds.json'
with open(output_file, 'w') as f:
    json.dump(data, f, indent=2)

final_land_count = len(data['objects']['land']['geometries'])

print(f"Original number of land features: {initial_land_count}")
print(f"Removed 1 feature (Antarctica).")
print(f"New number of land features: {final_land_count}")
print(f"Updated data saved to: '{output_file}'")
```
